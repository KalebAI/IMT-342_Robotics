A_i = 
\begin{bmatrix}
\cos\theta_i & -\sin\theta_i\cos\alpha_i & \sin\theta_i\sin\alpha_i & a_i\cos\theta_i \\
\sin\theta_i & \cos\theta_i\cos\alpha_i  & -\cos\theta_i\sin\alpha_i & a_i\sin\theta_i \\
0            & \sin\alpha_i              & \cos\alpha_i              & d_i \\
0            & 0                         & 0                         & 1
\end{bmatrix}

```python
import numpy as np
import sympy as sp
```

```python
# Funciones de nomenclatura
def C(ang):
    return sp.cos(ang)
def S(ang):
    return sp.sin(ang)

def A(th, d, a, al):
    M = sp.Matrix([[C(th), -S(th)*C(al), S(th)*S(al), a*C(th)],
                   [S(th), C(th)*C(al), -C(th)*S(al), a*S(th)],
                   [0, S(al), C(al), d],
                   [0, 0, 0, 1]])
    return M
```

```python
# Creo variables simbolicas
q1, q2, q3, q4 = sp.symbols('q_1 q_2 q_3 q_4')
th1, th2, th3, th4 = sp.symbols('theta_1 theta_2 theta_3 theta_4')
d1, d2, d3, d4 = sp.symbols('d_1 d_2 d_3 d_4')
a1, a2, a3, a4 = sp.symbols('a_1 a_2 a_3 a_4')
al1, al2, al3, al4 = sp.symbols('alpha_1 alpha_2 alpha_3 alpha_4')
```

```python
# Creo mis matrices de los eslabones

A01 = A(th1, d1, a1, al1)
A12 = A(th2, d2, a2, al2)
A23 = A(th3, d3, a3, al3)
A34 = A(th4, d4, a4, al4)

AT = A01*A12*A23*A34
#sp.pretty_print(A01)
#sp.pretty_print(A12)

AT
```
